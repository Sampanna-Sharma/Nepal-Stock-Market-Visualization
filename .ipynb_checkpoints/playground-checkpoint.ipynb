```python
import numpy as np
import pandas as pd
```

```python
processed_data = pd.read_csv('processed_data.csv')
```

```python
processed_data.head()
```

```python
from datetime import datetime
```

```python
dates = processed_data['time']
dates = dates.astype(int)
```

```python
dates = pd.to_datetime(dates, unit='ms')
```

```python
processed_data['time'] = dates
processed_data.head()
```

```python
processed_data.groupby('time').head(1)
```

```python
data_sample = processed_data[1:5]
```

```python
data_sample = data_sample.append(processed_data[10000:10005], ignore_index=True)
```

```python
data_sample = data_sample.append(processed_data[80000:80005], ignore_index=True)
```

```python
data_sample = data_sample.append(processed_data[1000:1005], ignore_index=True)
data_sample
```

```python
year_data = data_sample.groupby(data_sample['time'].dt.to_period('Y'))['high'].agg(['max'])
```

```python
year_data['min'] = data_sample.groupby(data_sample['time'].dt.to_period('Y'))['low'].agg(['min'])
```

```python
year_data['no_of_trans'] = data_sample.groupby(data_sample['time'].dt.to_period('Y'))['no_trans'].agg(['sum'])
```

```python
company_data = processed_data.loc[processed_data['traded_companies'] == 'Nabil Bank Limited']
```

```python
company_data_year = company_data.groupby(company_data['time'].dt.to_period('Y'))['high'].agg(['max'])
company_data_year['min'] = company_data.groupby(company_data['time'].dt.to_period('Y'))['low'].agg(['min'])
company_data_year['no_of_trans'] = company_data.groupby(company_data['time'].dt.to_period('Y'))['no_trans'].agg(['sum'])
```

```python
company_data_year
```

```python
company_data_year['open'] = company_data.groupby(company_data['time'].dt.to_period('Y'))['open'].agg(['first'])
```

```python
company_data_year['close'] = company_data.groupby(company_data['time'].dt.to_period('Y'))['close'].agg(['last'])
```

```python
company_data_year.loc[company_data_year.close > company_data_year.open, 'color'] = 'green'
company_data_year.loc[company_data_year.close <= company_data_year.open, 'color'] = 'red'
```
